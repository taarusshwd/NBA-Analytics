```python
# import libraries required

from urllib.request import urlopen
from bs4 import BeautifulSoup
import pandas as pd  
import numpy as np
import requests
import datetime
```

```python
df = pd.DataFrame(columns=["Name", "NBA", "Team", "Season", "GP", "GS", "MIN", "FG", "FG%", "3PT", "3P%", "FT", "FT%", "OR", "DR", "REB", "AST", "BLK", "STL", "PF", "TO", "PTS"])

url = "https://www.espn.in/nba/teams"
count = 0
response = requests.get(url)    
data = response.text

# create beautiful soup object from HTML
soup = BeautifulSoup(data, 'html.parser')

all_teams = soup.find('div', {'class':'layout is-split'})

divisions = all_teams.find_all('div', {'class':'mt7'})

for div in divisions: 
    teams = div.find_all('div', {'class':'pl3'})    
    for team in teams: 
        links = team.find_all('a')
        extension = links[3]['href']
        baseurl = 'https://www.espn.in'
        nexturl = baseurl + extension
        
        roster = requests.get(nexturl)
        rosterData = roster.text
        
        rosterSoup = BeautifulSoup(rosterData, 'html.parser')
        
        table = rosterSoup.find('table', {'class':'Table'})
        players = table.find_all('tr', {'class':"Table__TR Table__TR--lg Table__even"})
        for player in players:
            playerurl = player.find('a')['href']
            statsurl = playerurl[0:31] + 'stats' + playerurl[30:]
            temp = playerurl.split('/')[-1]
            n = temp.split('-')[0].capitalize() + ' ' + temp.split('-')[1].capitalize()
            
            stats = requests.get(statsurl)
            allStats = stats.text
        
            statsSoup = BeautifulSoup(allStats, 'html.parser')
            
            nbatable = statsSoup.find('div', {'class':'ResponsiveTable ResponsiveTable--fixed-left pt4'})
            if(nbatable):
                years = nbatable.find('tbody', {'class':"Table__TBODY"})
            else:
                continue
            
            seasons = years.find_all('tr', {'class':"Table__TR Table__TR--sm Table__even"})
            num = len(seasons)

            listSeasons = []
            teamsPlayed = []
            for season in seasons:
                impdata = season.find_all('td')
                listSeasons.append(impdata[0].text)
                teamsPlayed.append(impdata[1].text)
            
            numbers = nbatable.find('div', {'class':'Table__Scroller'})
            allstats = numbers.find('tbody', {'class':"Table__TBODY"})
            careerStats = allstats.find_all('tr')
            i = 0
            for yearStat in careerStats:
                eachstat = yearStat.find_all('td')
                s = listSeasons[i]
                t = teamsPlayed[i]
                i += 1
                gp = eachstat[0].text
                gs = eachstat[1].text
                mins = eachstat[2].text
                fg = eachstat[3].text
                fgp = eachstat[4].text
                threept = eachstat[5].text
                threeptp = eachstat[6].text
                ft = eachstat[7].text
                ftp = eachstat[8].text
                OR = eachstat[9].text
                dr = eachstat[10].text
                reb = eachstat[11].text
                ast = eachstat[12].text
                blk = eachstat[13].text
                stl = eachstat[14].text
                pf = eachstat[15].text
                to = eachstat[16].text
                pts = eachstat[17].text
            
                df = df.append({"Name": n, "NBA": 1, "Team": t, "Season": s, "GP": gp, "GS": gs, "MIN": mins, "FG": fg, "FG%": fgp, "3PT": threept, "3P%":threeptp, "FT":ft, "FT%":ftp, "OR":OR, "DR":dr, "REB":reb, "AST":ast, "BLK":blk, "STL":stl, "PF":pf, "TO":to, "PTS":pts}, ignore_index=True)
            
            statsurl = statsurl.rsplit('/', 1)[0] + "/type/mens-college-basketball" 
            
            stats = requests.get(statsurl)
            allStats = stats.text
        
            statsSoup = BeautifulSoup(allStats, 'html.parser')
            
            nbatable = statsSoup.find('div', {'class':'ResponsiveTable ResponsiveTable--fixed-left pt4'})
            if(nbatable):
                years = nbatable.find('tbody', {'class':"Table__TBODY"})
            else:
                continue
            
            seasons = years.find_all('tr', {'class':"Table__TR Table__TR--sm Table__even"})
            count = 0
            listSeasons = []
            teamsPlayed = []
            for season in seasons[:-1]:
                impdata = season.find_all('td')
                listSeasons.append(impdata[0].text)
                teamsPlayed.append(impdata[1].text)
            
            numbers = nbatable.find('div', {'class':'Table__Scroller'})
            allstats = numbers.find('tbody', {'class':"Table__TBODY"})
            careerStats = allstats.find_all('tr')
            i = 0
            for yearStat in careerStats[:-1]:
                eachstat = yearStat.find_all('td')
                s = listSeasons[i]
                t = teamsPlayed[i]
                i += 1
                gp = eachstat[0].text
                gs = eachstat[1].text
                mins = eachstat[2].text
                fg = eachstat[3].text
                fgp = eachstat[4].text
                threept = eachstat[5].text
                threeptp = eachstat[6].text
                ft = eachstat[7].text
                ftp = eachstat[8].text
                OR = eachstat[9].text
                dr = eachstat[10].text
                reb = eachstat[11].text
                ast = eachstat[12].text
                blk = eachstat[13].text
                stl = eachstat[14].text
                pf = eachstat[15].text
                to = eachstat[16].text
                pts = eachstat[17].text
            
                df = df.append({"Name": n, "NBA": 0, "Team": t, "Season": s, "GP": gp, "GS": gs, "MIN": mins, "FG": fg, "FG%": fgp, "3PT": threept, "3P%":threeptp, "FT":ft, "FT%":ftp, "OR":OR, "DR":dr, "REB":reb, "AST":ast, "BLK":blk, "STL":stl, "PF":pf, "TO":to, "PTS":pts}, ignore_index=True)         
```
